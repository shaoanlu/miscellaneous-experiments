## Load CIFAR-10

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# Training parameters
batch_size = 128  # orig paper trained all networks with batch_size=128
epochs = 200
data_augmentation = True
num_classes = 10

# Input image dimensions.
input_shape = x_train.shape[1:]

# Normalize data.
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
x_train_mean = np.mean(x_train, axis=0)
x_train -= x_train_mean
x_test -= x_train_mean

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print('y_train shape:', y_train.shape)

# Convert class vectors to binary class matrices.
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

## Build and compile model

```python
model = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)

model2 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)

model3 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)

model4 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)

model5 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)
model6 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)
model7 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)
model8 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)
model9 = ResNet50V2(
    include_top=True,
    weights=None,
    input_shape=(32, 32, 3),
    classes=10
)
```

```python
from tensorflow. keras.callbacks import ReduceLROnPlateau
lr_reducer = ReduceLROnPlateau(factor=np.sqrt(0.1),
                               cooldown=0,
                               patience=5,
                               min_lr=0.5e-6)

from tensorflow.keras.optimizers import Adam
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model2.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model3.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model4.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model5.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model6.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model7.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model8.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
model9.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=3e-4),
              metrics=['accuracy'])
```

## Build Data generator

```python
from tensorflow. keras.preprocessing.image import ImageDataGenerator
```

```python
def blur_input(x):
  ratio = np.random.uniform(0.1, 0.25)
  x = cv2.resize(x, (0, 0), fx=ratio, fy=ratio)
  x = cv2.resize(x, (32, 32))
  return x

def random_blur_input(x):
  if np.random.choice([True, False]):
    ratio = np.random.uniform(0.1, 0.25)
    x = cv2.resize(x, (0, 0), fx=ratio, fy=ratio)
    x = cv2.resize(x, (32, 32))
  return x
```

```python
datagen1 = ImageDataGenerator(
    preprocessing_function=blur_input,
    horizontal_flip=True,
)

datagen2 = ImageDataGenerator(
    horizontal_flip=True,
)

datagen3 = ImageDataGenerator(
    preprocessing_function=random_blur_input,
    horizontal_flip=True,
)
```

```python
datagen1.fit(x_train)
datagen2.fit(x_train)
datagen3.fit(x_train)
```

## Start training

```python
callbacks = [lr_reducer]
```
